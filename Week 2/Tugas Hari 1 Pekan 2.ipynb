<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

>**Jawab:**<br>
- Perbedaan Series dengan List:
  - Perbedaan method yang mendukung.
  - Pada Series terdapat index dan value sedangkan pada List hanya value saja. Index pada Series juga dapat di-kustom menggunakan String sedangkan pada List secara sistem hanya menggunakan nilai integer sebagai index yang dimulai 0.
- Perbedaan Series dengan Dictionary:
  - Perbedaan method yang mendukung.
  - Pada Series dapat memiliki key yang yang non-unique sedangkan Dictionary harus unique.
  - Pada Series dapat dilakukan proses slicing sedangkan pada Dictionary tidak bisa.
- Perbedaan Series dengan DataFrame:
  - Series merupakan struktur dasar pada Pandas. Ibaratnya seperti array 1 dimensi yang memiliki index.
  - DataFrame memiliki struktur 2 dimensi dengan baris dan kolom. Ibaratnya sebuah DataFrame terdiri dari lebih dari 1 Series.

```python
# code here
pd.Series([3.77, 3.90], index=['ivan', 'ande'])
```
