<h2>Soal 2: Aritmatika di Python </h2>
    
- Buat statement pertambahan antara 2 number di Python
- Buat statement perkurangan antara 2 number di Python
- Buat statement perkalian antara 2 number di Python
- Buat statement pembagian antara 2 number di Python

```python
bilangan1 = 12
bilangan2 = 5

operasi_pertambahan = bilangan1 + bilangan2
operasi_pengurangan = bilangan1 - bilangan2
operasi_perkalian = bilangan1 * bilangan2
operasi_pembagian = bilangan1 / bilangan2

print(f'{bilangan1} + {bilangan2} = {operasi_pertambahan}')
print(f'{bilangan1} - {bilangan2} = {operasi_pengurangan}')
print(f'{bilangan1} x {bilangan2} = {operasi_perkalian}')
print(f'{bilangan1} : {bilangan2} = {operasi_pembagian}')
```
